```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

```python
#Line Charts (Ã‡izgi Grafikleri)
```

```python
col1 = np.linspace(0, 10, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({"C1" : col1, "C2" : col2})
df.head(10)
```

```python
plt.style.use('seaborn-darkgrid')
%matplotlib inline
sns.lineplot(x = df.C1, y = df.C2, data = df)
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.set(rc = {"axes.facecolor" : "#FFFDE7", "axes.grid" : True, 'xtick.labelsize' : 14, 'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 2)
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 2)
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid" : False, 'xtick.labelsize' : 14, 'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, color = "#FF5722", linewidth = 2)
sns.lineplot(x = df.C1, y = np.cos(df.C1), data = df, color = "#FFEB3B", linewidth = 2)
plt.show()
```

```python
canada = pd.read_csv('mydata/canada1.csv', delimiter = ";")
canada.head()
```

```python
canada.columns
```

```python
canada.drop(columns = ['AREA', 'DEV', 'DevName', 'REG', 'Type', 'Coverage', 'AreaName', 'RegName'], inplace = True)
canada.head()
```

```python
canada.rename(columns = {'OdName':'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
```

```python
canada.head()
```

```python
plt.figure(figsize = (14,6))
plt.title("Immigrants from Turkey", fontsize = 14)
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid" : False, 'xtick.labelsize' : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Turkey'])
plt.show()
```

```python
plt.figure(figsize = (16,6))
plt.title("Immigrants", fontsize = 14)
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid" : False, 'xtick.labelsize' : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Ukraine'], label = 'Ukraine')
sns.lineplot(x = canada.index.values, y = canada['Viet Nam'], label = 'Vietnam')
sns.lineplot(x = canada.index.values, y = canada['Syrian Arab Republic'], label = 'Syria')
sns.lineplot(x = canada.index.values, y = canada['Poland'], label = 'Poland')
plt.legend(facecolor = 'grey', fontsize = 'large', edgecolor = 'black', shadow = True)
plt.show()
```

```python
pokemon = pd.read_csv('mydata/pokemon.csv')
pokemon.head()
```

```python
pokemon.drop(columns = ['#'], inplace = True)
pokemon.head()
```

```python
plt.figure(figsize = (20, 7))
sns.barplot(x = pokemon['Type 1'], y = pokemon['Attack'])
plt.show()
```

```python
plt.figure(figsize = (20, 7))
sns.barplot(x = pokemon['Type 1'], y = pokemon['Defense'])
plt.show()
```

```python
#sorted bar plot
```

```python
plt.figure(figsize = (20, 7))
order = pokemon.groupby(['Type 1']).mean().sort_values('Defense', ascending = False).index
sns.barplot(x = pokemon['Type 1'], y = pokemon['Defense'], order = order)
plt.show()
```

```python
plt.figure(figsize = (20, 7))
sns.barplot(x = pokemon['Type 1'], y = pokemon['Speed'])
plt.show()
```

```python
plt.figure(figsize = (8, 10))
sns.barplot(x = pokemon['Speed'], y = pokemon['Type 1'])
plt.show()
```

```python
plt.figure(figsize = (20, 7))
sns.countplot(x = pokemon['Type 1'])
plt.show()
```

```python
plt.figure(figsize = (20, 7))
sns.countplot(x = pokemon['Type 1'], order = pokemon['Type 1'].value_counts().index)
plt.show()
```
