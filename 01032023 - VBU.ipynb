```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

```python
pokemon = pd.read_csv('mydata/pokemon.csv')
pokemon.head()
```

```python
sns.catplot(y = "Type 1", kind = "count", data = pokemon)
plt.show()
```

```python
sns.catplot(y = "Type 1", kind = "count", data = pokemon, order = pokemon['Type 1'].value_counts().index)
plt.show()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
plt.figure(figsize = (7,7))
sns.catplot(x = 'species', y = 'petal_width', kind = 'violin', data = iris)
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize = (7,7))
sns.catplot(x = 'petal_width', y = 'species', kind = 'violin', data = iris, height = 5)
plt.show()
```

```python
fig1, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (30,30))
sns.swarmplot(x = "species", y = "petal_width", ax = axes[0, 0], data = iris, size = 8)
sns.swarmplot(x = "species", y = "petal_length", ax = axes[0, 1], data = iris, size = 8)
sns.swarmplot(x = "species", y = "sepal_width", ax = axes[1, 0], data = iris, size = 8)
sns.swarmplot(x = "species", y = "sepal_length", ax = axes[1, 1], data = iris, size = 8)
plt.show()
```

```python
#Pair Plot
```

```python
fish = pd.read_csv('mydata/Fish.csv')
fish.head()
```

```python
fish1 = fish[fish['Species'].isin(['Bream', 'Perch', 'Pike'])]
fish1.head()
```

```python
plt.figure(figsize = (11,9))
sns.pairplot(fish1, hue = "Species")
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', palette = "husl", size = 2)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Length1", "Weight", "Height", "Width"])
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Length1", "Weight", "Height", "Width"], markers = ['o', 'D', '<'])
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], markers = ['o', 'D', '<'], height = 5)
plt.show()
```

```python
#x_vars ve y_vars parametreleri
```

```python
sns.pairplot(fish1, hue = "Species", vars = ["Weight", "Height", "Width"], kind = "reg", height = 5)
plt.show()
```

```python
sns.pairplot(fish1, hue = "Species", vars = ["Weight", "Height", "Width"], plot_kws = dict(s = 140), height = 5)
plt.show()
```

```python
sns.pairplot(fish1, hue = "Species", vars = ["Weight", "Height", "Width"], plot_kws = dict(s = 140, alpha = 0.6), height = 5)
plt.show()
```

```python
sns.pairplot(fish1, hue = "Species", x_vars = ["Weight", "Height"], y_vars = ['Weight', 'Height'],
             plot_kws = dict(s = 140, linewidth = 1, alpha = 0.6), height = 5)
plt.show()
```

```python
g = sns.PairGrid(fish1)
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species')
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
g = g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species', palette = "husl")
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
g = g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species', palette = "husl")
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist, histtype = 'step', linewidth = 2)
g = g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"])
g = g.map_offdiag(plt.scatter, edgecolor = "w", s = 130)
g = g.map_diag(plt.hist, edgecolor = "w", linewidth = 2)
g = g.add_legend()
plt.show()
```

```python
plt.plot(figsize = (9,9))
sns.regplot(x = fish1.Weight, y = fish1.Height, color = '#FF6600')
plt.show()
```
